```python
def scatter_plotting(x_axis, y_axis):
    x = x_axis
    y = y_axis
    
    fig,ax = plt.subplots(figsize=(8, 6))
    scatter = ax.scatter(x, y, color='dodgerblue', edgecolor='white', s=100, alpha=0.1)
    ax.scatter(x, y)

    ax.set(xlim=(0, 190),
           ylim=(0, 230))
    # x = df['Attack'].max()
    # 190
    # y = df['Defense'].max()
    # 230
    
    ax.set_xlabel('Attack')
    ax.set_ylabel('Defense')
    ax.set_title('Pokemon Attack vs Defense')
    

    m,b = np.polyfit(x,y,1)
    # plt.plot(xlabel = "Attack")
    # plt.plot(ylabel = "Defense")
    plt.plot(x, m*x+b, c = 'black')
    plt.show()

scatter_plotting(df['Attack'],df['Defense'])
```
